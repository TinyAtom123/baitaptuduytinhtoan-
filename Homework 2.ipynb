```python
#Bai 8


from sympy import Eq, solve, symbols
a, b, c, d = map(float,input().split())
m, n = map(float,input().split())
x, y = symbols("x, y")
if a / b == c / d:
    print("He phuong trinh vo so nghiem")
elif a / b == c / d != m / n:
    print("He phuong trinh vo nghiem")
else: 
   eq1 = Eq(a*x + b*y, m)                       # 2 phuong trinh 
   eq2 = Eq(c*x + d*y, n)
   sol = solve((eq1, eq2), (x, y))              #Cho nay tra sol[x] va sol[y] 
   x1 = round(float(sol[x]), 2)
   y2 = round(float(sol[y]), 2)
   print(x1, y2)

```

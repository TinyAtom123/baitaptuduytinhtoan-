```python
#Bai 4


h = int(input())
i = 1
while i <= h:
    j = 1
    while j <= 2*h - i:
        if j == h - i + 1 or j == h + i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="")
        j += 1
    print()
    i += 1
```

```python
#Bai 20


import math


n = int(input())
lst = []
for i in str(n):
    i = int(i)
    is_prime = True
    if i < 2:
        is_prime = False
    for j in range(2, int(math.sqrt(i)) + 1):
        if i % j == 0:
            is_prime = False
        else:
            is_prime = True
    if is_prime:
        lst.append(i)
   
    

print("So cac so nguyen to trong so", n, "la:", len(lst))
```

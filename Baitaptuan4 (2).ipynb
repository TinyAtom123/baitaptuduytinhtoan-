```python
#W4A13


a = int(input())
b = int(input())
lst =[]
kjh = []
for i in range(1, a):
    if a % i == 0:
        lst.append(i)
for j in range(1, b):
    if b % j == 0:
        kjh.append(j)
if sum(lst) == b and sum(kjh) == a:
    print("true")
else:
    print("false")
```

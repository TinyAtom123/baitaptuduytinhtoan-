```python
#W3A13


n = int(input())
tong = 0
if 0 <= n <= 50:
    tong = n*1500
    print(tong)
elif 51 <= n <= 100:
    tong = 50*1500 + (n-50)*2000
    print(tong)
elif n > 100:
    tong = 50*1500 + 50*2000 + (n-100)*3000
    print(tong)
else:
    print("So khong hop le")
```
